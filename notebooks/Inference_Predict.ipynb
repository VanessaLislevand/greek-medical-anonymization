# Inference Predict

Notebook for running anonymization on:

- a single `.docx` / `.txt` report
- a directory of reports for batch processing


## 1. Setup

Run this notebook from the project environment after installing the package with:

```bash
pip install -e '.[ml]'
```

````python
from pathlib import Path
import json

from greek_med_anonymizer.cli import run_anonymize
````

## 2. Paths

Set the project root, config file, and either a single input file or an input directory.

````python
PROJECT_ROOT = Path("/Users/vanessalislevand/Documents/New project")

# Choose one of the provided config files and update model_dir inside it beforehand.
CONFIG_PATH = PROJECT_ROOT / "examples" / "config.free_text_only.example.json"

# Single-file mode
INPUT_FILE = Path("/absolute/path/to/report.docx")
OUTPUT_FILE = Path("/absolute/path/to/report.anon.txt")

# Batch mode
INPUT_DIR = Path("/absolute/path/to/reports_directory")
OUTPUT_DIR = Path("/absolute/path/to/anonymized_directory")
````

## 3. Inspect config

Optional sanity check before running inference.

````python
config_preview = json.loads(CONFIG_PATH.read_text(encoding="utf-8"))
config_preview
````

## 4. Single-file inference

````python
run_anonymize(
    input_path=str(INPUT_FILE),
    output_path=str(OUTPUT_FILE),
    config_path=str(CONFIG_PATH),
    emit_metadata=True,
)

print("Saved anonymized text to:", OUTPUT_FILE)
print("Saved metadata to:", Path(str(OUTPUT_FILE) + ".json"))
````

## 5. Preview single-file output

````python
print(OUTPUT_FILE.read_text(encoding="utf-8")[:4000])
````

## 6. Batch inference

````python
run_anonymize(
    input_path=str(INPUT_DIR),
    output_path=str(OUTPUT_DIR),
    config_path=str(CONFIG_PATH),
    emit_metadata=True,
)

print("Saved batch outputs under:", OUTPUT_DIR)
````

## 7. List batch outputs

````python
sorted(str(path) for path in OUTPUT_DIR.rglob("*") if path.is_file())[:50]
````
